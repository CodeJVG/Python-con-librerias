# Quick Plotting



```python
#Importamos las libreias
import numpy as np
import matplotlib.pyplot as plt
```


## Line Plots
Comando simple para obtener data rapidamente

```python
x = np.linspace(0,15,30)
y = np.sin(x) + 0.1*np.random.randn(len(x))
```

```python
plt.grid() #Para la cuadricula
plt.plot(x,y,lw= 2)
```

```python
plt.plot(x,y,'--')
```

```python
plt.plot(x,y,'o--')
```

```python
plt.plot(x,y,'o--', color = 'purple', lw=1.7, ms=10)
```

Tamaño de figura

```python
plt.figure(figsize=(10,3))
plt.plot(x,y,'o--', color = 'purple', lw=1.7, ms=10)
```

Etiqueta de ejes

```python
plt.figure(figsize=(10,3))
plt.plot(x,y, 'o--', color='green',lw=0.7,ms=5)
plt.xlabel("Time [s]")
plt.ylabel("Voltaje [s]")
```

Leyenda

```python
plt.figure(figsize=(10,3))
plt.plot(x,y,'o--',color='green', lw=.7, ms=5, label='Componente 1')
plt.xlabel("Time [s]")
plt.ylabel("Voltaje [s]")
plt.legend()
```

Para varias graficas

```python
x2 = np.linspace(0,15,100)
y2 = np.sin(x2)

plt.figure(figsize=(10,3))
plt.plot(x,y,'o',lw=0.7,ms=5,label="Componenete 1")
plt.plot(x2,y2,'--',lw=0.7,ms=5, label='componente 2')
plt.legend()
```

Limites en X y Y

```python
x2 = np.linspace(0,15,100)
y2 = np.sin(x2)

plt.figure(figsize=(10,3))
plt.plot(x,y,'o',lw=0.7,ms=5,label="Componenete 1")
plt.plot(x2,y2,'--',lw=0.7,ms=5, label='componente 2')
plt.xlabel("Time [s]")
plt.ylabel("Voltaje [s]")
plt.ylim(top=1.5)
plt.legend(loc="upper right", fontsize=10,ncol=2)
plt.grid()
```

```python
x2 = np.linspace(0,15,100)
y2 = np.sin(x2)

plt.figure(figsize=(10,3))
plt.plot(x,y,'*',color="orange",ms=5,label="Componenete 1")
plt.plot(x2,y2,'-',lw=0.7,ms=5, label='Componente 2')
plt.xlabel("Time [s]")
plt.ylabel("Voltaje [s]")
plt.ylim(top=2)
plt.xlim(left=-1)
plt.legend(loc="lower left", fontsize=10)
plt.grid()
```

## Un eje por figura

```python
fig, ax = plt.subplots(1,1, figsize=(12,4))
ax.plot(x, y, "o--")
ax.set_xlabel('eje x')
ax.set_ylabel('eje y')
```

## Multiples ejes por figura

```python
fig, axes = plt.subplots(2,2, figsize=(12,10))
ax = axes[0][0]
ax.plot(x, y, "o--")
ax.set_xlabel('eje x')
ax.set_ylabel('eje y')
ax.grid()

ax.text(0.1,0.1,"Texto aqui", transform=ax.transAxes) #Para colocar texto
ax.tick_params(axis="both", labelsize=15) #Tamaño de letra en los ejes


ax= axes[1][1]
ax.hist(res, bins=30,density=True, histtype='step', label='componente 1')
ax.set_xlabel('eje x')
ax.set_ylabel('eje y')
ax.legend()
ax.set_title('Histograma')
plt.show()
```

## Ejemplos


```python
res_a1 = 0.2*np.random.randn(1000)*0.4
res_b1 = 0.25*np.random.randn(1000)*0.4
res_a2 = 0.21*np.random.randn(1000)*0.3
res_b2 = 0.21*np.random.randn(1000)*0.3
```

```python
textstr1 = '\n'.join((
    r'$\sigma_a-%.4f$' % (np.std(res_a1)),
    r'$\sigma_b-%.4f$' % (np.std(res_b1))))
textstr2 = '\n'.join((
    r'$\sigma_a-%.4f$' % (np.std(res_a2)),
    r'$\sigma_b-%.4f$' % (np.std(res_b2))))
```

```python
textstr1
```

```python
fig, axes = plt.subplots(1,2, figsize=(10,3.5))
ax = axes[0]
ax.hist(res_a1,bins=30, density=True, histtype='step', label='a')
ax.hist(res_b1,bins=30, density=True, histtype='step',  label='b')
ax.legend(loc="upper right")

#Para poner un texto en una caja
ax.text(0.05,0.81,textstr1, transform=ax.transAxes, bbox=dict(facecolor="white", edgecolor="black"))
ax.set_ylabel("Density",fontsize=15)
ax.set_title("Histograma 1")

ax = axes[1]
ax.hist(res_a2,bins=30, density=True, histtype='step', label='a')
ax.hist(res_b2,bins=30, density=True, histtype='step',  label='b')
ax.legend(loc="upper right")
ax.text(0.05,0.81,textstr2, transform=ax.transAxes, bbox=dict(facecolor="white", edgecolor="black"))
ax.set_ylabel("Density",fontsize=15)
ax.set_title("Histograma 2")


fig.text(0.5,-0.04, "$\Delta E[Joules]$ ",ha="center", fontsize=20)#horizontal alifned
plt.show()
```

## Graficos en 2D

```python
#Creamos data
_ = np.linspace(-1,1,100)
x , y = np.meshgrid(_,_)
z = x**2+x*y
```

```python
z
```

### Diagramas de relleno de contorno  plt.contourf()

```python
#Mapa de calor
plt.contourf(x,y,z, levels=30,vmin=-0.2,vmax=1.68,cmap="plasma") #Vamos modificando los niveles
#vmin para el valor con el color mas oscuro
#vmax para el valor mas claro 
plt.colorbar(label="Tabla de color")#Insertamos una tabla de color

plt.figure(figsize=(10,15))
plt.show()
```

Para mas colores y estilos vease: https://matplotlib.org/3.5.0/tutorials/colors/colormaps.html

Tambien vease la documentación de este tipo de grafico : https://matplotlib.org/3.5.0/api/_as_gen/matplotlib.pyplot.contourf.html


### Solo contorno plt.contour()

```python
cs = plt.contour(x,y,z, levels=20)
plt.clabel(cs, fontsize=8) #Etiqueta a las curvas de nivel, indicar grafico
```

# Graficas 3D

```python
fig, ax = plt.subplots(subplot_kw={'projection':'3d'})
ax.plot_surface(x,y,z,cmap="coolwarm")
```

```python
#Creamos data
_ = np.linspace(-2,2,100)
x , y = np.meshgrid(_,_)
z1 = np.sin(x)+2*np.sin(y)
z2 = np.sqrt(3)/2*x + np.sqrt(3)*y - 2/3 - np.pi/4*np.sqrt(3)
fig, ax = plt.subplots(subplot_kw={'projection':'3d'})
ax.plot_surface(x,y,z1,cmap="coolwarm")
ax.plot_surface(x,y,z2,cmap="coolwarm")
```

# Graficas de flujo

```python
w = 3
_ = np.linspace(-3,3,100)
X , Y = np.meshgrid(_,_)
U = -1 - X**2 + Y
V = 1 +X - Y**2
speed = np.sqrt(U**2 + V**2)
```

```python
fig, axes = plt.subplots(2,2,figsize=(7,7))
ax = axes[0][0]
ax.streamplot(X,Y,U,V)

ax = axes[0][1]
ax.streamplot(X,Y,U,V,color=speed)

ax = axes[1][0]
lw = 5*speed /speed.max()
ax.streamplot(X,Y,U,V, linewidth= lw)

ax = axes[1][1]
lw = 5*speed /speed.max()
seedpoints= np.array([[0,1],[1,0]])
ax.streamplot(X,Y,U,V, start_points=seedpoints)
```
